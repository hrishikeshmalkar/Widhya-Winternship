### Microtask 2 : Feature Selection and Data Cleaning

#### Selecting required features

```python
features = ['keywords', 'cast', 'genres', 'director']
for feature in features:
    df[feature] = df[feature].fillna('')
```

#### creating Function: 'combined_features'

```python
def combined_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

```python
df["combined_features"] = df.apply(combined_features, axis =1)
```

#### preview after adding new feature

```python
df.head()
```

### Microtask 3 : Find Similiraity

#### Vectorizing data

```python
cv=CountVectorizer()
count_vector=cv.fit_transform(df['combined_features'])
```

```python
print("Count Vector:", count_vector.toarray())
```

#### Identifying Similarities

```python
x=[]
y=0
for i in count_vector:
    x.append([cosine_similarity(i,count_vector[0]),y])
    y+=1
```

```python
x[0]
```

```python
cosine_sim = cosine_similarity(count_vector)
print(cosine_sim)
```

```python
def similar_movies_identifier():
    
    movie=input('Please Enter Movie Title:')
    
    def get_index_from_title(title):
        return df[df.title == title]["index"].values[0]

    movie_index = get_index_from_title(movie)
    
    def get_title_from_index(index):
        return df[df.index == index]["title"].values[0]

    movie_title = get_title_from_index(index_movie)
    
    similar_movies = list(enumerate(cosine_sim[movie_index]))

    similar_movies_after_sort = sorted(similar_movies, key=lambda x:x[1], reverse=True)
    
    print("Top 10 similar movies to selected movie:", movie, "\n")

    i=0
    for movie in similar_movies_after_sort:
        print(i + 1, get_title_from_index(movie[0]))
        i=i+1
        if i>10:
            break
```
